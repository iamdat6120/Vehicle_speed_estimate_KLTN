```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
lst = []
with open('a.text','r') as f:
    # a = f.read()
    # print(a.split('\n'))
    for line in f.readlines():
        # print(float(line))
        lst.append(float(line))
```

```python
np.mean(lst)
```

```python
plt.plot(, 'o')




```

```python
pd.DataFrame(np.array[range(0, len(lst)),lst])
```

```python
df = pd.DataFrame(np.concatenate((np.array(range(0,len(lst))).reshape(-1,1), np.array(lst).reshape(-1,1)), axis = 1), columns =['a', 'b'])
```

```python
df
```

```python
import seaborn as sns

# Vẽ biểu đồ swarm
plt.figure(figsize=(16, 8))
sns.violinplot(x='a', y='b', data = df)
plt.xlabel('Scale', fontsize=16)
plt.ylabel('cm', fontsize=16)
plt.title("Swarnplot of Sepal.Width", fontsize=18)
```

```python
import seaborn as sns

# Vẽ biểu đồ swarm
plt.figure(figsize=(16, 8))
sns.histplot(x='b', data = df, bins=50, kde=True)
plt.xlabel('Scale', fontsize=16)
plt.ylabel('cm', fontsize=16)
plt.title("Histogram of frame rate", fontsize=18)
```
